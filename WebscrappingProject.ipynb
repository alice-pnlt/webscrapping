## Classement meilleures villes étudiantes françaises (L'Etudiant)

```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

driver = webdriver.Chrome(executable_path=r"/Users/paul/chromedriver_mac64-2/chromedriver")
driver.get("https://www.letudiant.fr/classements/classement-des-villes-etudiantes.html")

data = []

# Code pour enlever le pop-up de départ en cliquant sur "accepter"
try:
    element = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.XPATH, '//*[@id="sd-cmp"]/div[2]/div/div/div/div/div/div/div[2]/div[1]/button[1]'))
    )
    element.click()
except:
    pass

elements = driver.find_elements_by_xpath('/html/body/div[6]/div[2]/div[2]/div[1]/section[3]/div[2]/table/tbody/tr/td[3]/div/a')
data += [{'ville': element.text} for element in elements]

# Boucle pour parcourir les pages 2 et 4
for i in range(3,5):
    try:
        element = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, f'/html/body/div[6]/div[2]/div[2]/div[1]/section[3]/div[2]/section/div[2]/div/ul/li[{i}]/a'))
        )
        element.click()
    except:
        pass
    elements = driver.find_elements_by_xpath('/html/body/div[6]/div[2]/div[2]/div[1]/section[3]/div[2]/table/tbody/tr/td[3]/div/a')
    data += [{'ville': element.text} for element in elements]

df = pd.DataFrame.from_records(data)
df["Rank"] = df.reset_index().index + 1

print(df)
```

## Classement des meilleurs universités françaises (PlaneteGrandesEcoles)

```python
import requests
from lxml import html
import pandas as pd

url = "https://www.planetegrandesecoles.com/the-classement-2022-des-meilleurs-universites-francaises?fbclid=IwAR22RgwHylkxftRahwedaezrvLE9xNzW82iY34VTYxILO_N7-H7F0A1wb2o"
page = requests.get(url)
tree = html.fromstring(page.content)

xpath = '//*[@id="tablepress-86"]/tbody/tr'
rows = tree.xpath(xpath)

data = []
for row in rows:
    cols = row.xpath('.//td/text()')
    data.append([col.strip() for col in cols])

df2 = pd.DataFrame(data, columns=['Classement Mondial', 'Classement France', 'Université', 'Ville'])
print(df2)
```

## Classement prix du m2 en France (studyrama)

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd

driver = webdriver.Chrome(executable_path=r"/Users/paul/chromedriver_mac64-2/chromedriver")
driver.get("https://www.studyrama.com/vie-etudiante/les-dernieres-news-du-monde-etudiant/classement-des-villes-les-plus-cheres-de-france-109766?fbclid=IwAR1zANxa9A_sZdukphW8glgFiOP_T9CnBWbWbuq4Z63QbeX6oYf0ETBHWjc")

data = []

# Code pour enlever le pop-up de départ en cliquant sur "accepter"
try:
    element = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.XPATH, '//*[@id="sd-cmp"]/div[2]/div[1]/div/div/div/div/div/div[2]/div[1]/button[1]'))
    )
    element.click()
except:
    pass

villes = driver.find_elements_by_xpath('//*[@id="block-studyrama-content"]/div/div[6]/article/div/div[2]/ul[4]/li/table/tbody/tr/td[2]')
prix = driver.find_elements_by_xpath('//*[@id="block-studyrama-content"]/div/div[6]/article/div/div[2]/ul[4]/li/table/tbody/tr/td[3]')

for i in range(len(villes)):
    data.append({'Ville': villes[i].text, 'prix': prix[i].text})

df3 = pd.DataFrame(data)
```
